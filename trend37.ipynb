```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
df=pd.read_csv("/content/time.csv")
# Example: Generate a large dataset
np.random.seed(42)
# Creating a dataset with 1000 points
data_size = 1000
time = np.arange(data_size)  # Time series
values = np.random.normal(0, 1, data_size).cumsum() + np.arange(data_size) * 0.5  # Simulating a trend

# Load data into a DataFrame
df = pd.DataFrame({'Time': time , 'Value': values})

# Preparing the data for linear regression
X = df['Time'].values.reshape(-1, 1)  # Independent variable (Time)
y = df['Value'].values  # Dependent variable (Value)

# Apply linear regression
model = LinearRegression()
model.fit(X, y)

# Predict the trend
trend = model.predict(X)

# Calculate the slope (i.e., the rate of change)
slope = model.coef_[0]
intercept = model.intercept_

# Print trend information
print(f"Slope (Trend): {slope}")
print(f"Intercept: {intercept}")

# Plotting the data with the trend line
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], df['Value'], label='Data')
plt.plot(df['Time'], trend, color='red', label='Trend', linewidth=2)
plt.title(f'Trend Analysis (Slope: {slope:.2f})')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```
